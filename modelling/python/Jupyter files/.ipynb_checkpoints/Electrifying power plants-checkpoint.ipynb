```python
import numpy as np
P = 300
P_max = 600

A = -0.1308
B = 0.2426
C = 0.3398

Q_exact = []
Q_approx = []

def eta(X):
    return(A*X**2 + B*X + C)

for i,P in enumerate(np.arange(10,P_max,10)):
    X = P/P_max
    Q_exact.append (0.5 * ( X*P_max / eta(X) ) )
    Q_approx.append ( P_max*0.5* (0.962 +
                           2.085*(X-0.4))
                    )
                        
Fig = go.Figure()
Fig.add_trace(go.Scatter(x=np.arange(10,P_max,10), y=Q_exact, name='Exact'
                        ))
Fig.add_trace(go.Scatter(x=np.arange(10,P_max,10),y=Q_approx, name='Linearaised'
                        ))

Fig.update_xaxes({'title':'Pout [MW]',
                  'showgrid':True,
                  'gridcolor':'lightgray',
                  'linecolor':'black',
                  'mirror':True,
                  'showline':True})
Fig.update_yaxes({'title':'-dQ/0.5hr',
                  'ticks':'outside',
                 'nticks':8,
                  'showgrid':True,
                  'gridcolor':'lightgray',
                  'linecolor':'black',
                  'mirror':True,
                  'showline':True})

Fig.update_layout(plot_bgcolor='white',
                  width=600, height=450,
                  margin=dict(l=10, r=10, t=20, b=10),
                  legend=dict(
                              orientation="v",
                              yanchor="top",
                              y=0.95,
                              xanchor="center",
                              x=0.2,
                              )
                  )
Fig.show()
            
```

# Plot generation in SA

```python
import sys, os
print("Operating system: %s" %sys.platform)
if sys.platform == 'win32':
    curr_path = 'c:\\Nextcloud\\Thermal Battery Research\\modelling\\python'
elif sys.platform == 'linux':
    curr_path = '/home/jeff/cloud/documents/work/ANU/Thermal Battery Research/modelling/python'
else: print("What operating system are you running?! I've never even heard of %s" %sys.platform)
if curr_path not in sys.path:
    sys.path.append(curr_path)
    print('Path added! \n')
os.chdir(curr_path)
print("Working directory is now %s" %curr_path)
from projdirs import datadir
import pandas as pd
import package.sql_manager as sm

DT1 = '2020-01-1 22:00:00'
DT2 = '2020-01-20 22:00:00'


db_interchange = '\\NEM generation\\interchange_SA.db'
db_gen = '\\NEM generation\\generation_SA.db'
db_pv = '\\NEM generation\\rooftop_PV_generation.db'

generation_SA = sm.get_data('N/A','N/A',db_gen)
generation_SA.date_time = pd.to_datetime(generation_SA.date_time)

interchange_SA = sm.get_data('N/A','N/A',db_interchange)
interchange_SA.date_time = pd.to_datetime(interchange_SA.date_time)

PV = sm.get_data('N/A','N/A',db_pv)
PV_SA = PV[PV.region == 'SA1'].copy()
PV_SA.date_time = pd.to_datetime(PV.date_time)


Generation_SA = generation_SA.set_index('date_time').loc[DT1:DT2]
Interchange_SA = interchange_SA.set_index('date_time').loc[DT1:DT2]
PV_SA = PV_SA.set_index('date_time').loc[DT1:DT2]
PV_SA = PV_SA.resample('300S').interpolate(method='linear')

demand_total = -Interchange_SA.power + Generation_SA.drop('Misc',axis=1).sum(axis=1) + PV_SA.power_sattellite


import plotly.graph_objects as go
from plotly.subplots import make_subplots
Fig = go.Figure()

colors = {'Fossil':'saddlebrown',
          'Wind': 'green',
          'Solar': 'orange',
          'Battery': 'deepskyblue',
          'Waste': 'red'}

X = {'Fossil':1, 'Waste':1, 'Battery':1, 'Wind':1, 'Solar':1,'PV':1}

Interchange_new = -Interchange_SA.power\
                  - (X['PV']-1)*PV_SA.power_sattellite \
                  - (X['Solar']-1)*Generation_SA.Solar \
                  - (X['Wind']-1)*Generation_SA.Wind

Fig.add_trace(go.Scatter(x=Interchange_new.index,
                         y=Interchange_new,
                         line_shape = 'hv',
                         line = dict(width=0),
                         name='Import',
                         stackgroup='one'))


for type in ['Fossil', 'Waste', 'Battery', 'Wind', 'Solar']:
    if type in Generation_SA.columns: 
        Fig.add_trace(go.Scatter(x=Generation_SA.index,
                                 y=X[type]*Generation_SA[type],
                                 line_shape = 'hv',
                                 fill='tonexty',
                                 fillcolor=colors[type],
                                 opacity=0.5,
                                 mode='none',
                                 name=type,
                                 stackgroup='one'
                                ))

Fig.add_trace(go.Scatter(x=PV_SA.index,
                         y=X['PV']*PV_SA.power_sattellite,
                         line_shape = 'hv',
                         fill='tonexty',
                         fillcolor='yellow',
                         opacity=0.5,
                         mode='none',
                         name='Rooftop PV',
                         stackgroup='one'
                        ))

Fig.add_trace(go.Scatter(x = demand_total.index,
                         y= demand_total,
                         line_shape = 'hv',
                         line=dict(width=1,color='black'),
                         name='Demand'))

Fig.show()


```

# Plot the operational demand of SA from AEMO's Integrated System Planning

```python
import sys, os
print("Operating system: %s" %sys.platform)
if sys.platform == 'win32':
    curr_path = 'c:\\Nextcloud\\Thermal Battery Research\\modelling\\python'
elif sys.platform == 'linux':
    curr_path = '/home/jeff/cloud/documents/work/ANU/Thermal Battery Research/modelling/python'
else: print("What operating system are you running?! I've never even heard of %s" %sys.platform)
if curr_path not in sys.path:
    sys.path.append(curr_path)
    print('Path added! \n')
os.chdir(curr_path)
print("Working directory is now %s" %curr_path)
from projdirs import datadir
import pandas as pd
import package.sql_manager as sm
import numpy as np

def fix_data(data):
    data['date'] = pd.to_datetime(data[['Year','Month','Day']])
    data = data.drop(['Year','Month','Day'],axis=1).set_index('date')
    data = data.stack().reset_index(name='value').drop('level_1',axis=1)
    data['time'] = np.resize(np.arange(1,49,1)/2,data.shape[0])
    data['date_time'] = data.date+pd.to_timedelta(data['time'],unit='h')
    data.drop(['date','time'],axis=1,inplace=True)
    return data
    

path = 'C:\\Users\\z3337922\Downloads\\SA 2019 Draft Demand Traces by components - High DER\\SA - High DER\\'

demand_file = 'SA_RefYear_2019_HIGH_DER_POE10_OPSO.csv'
rooftopPV_file = 'SA_RefYear_2019_HIGH_DER_POE10_PV.csv'

demand = pd.read_csv(path+demand_file,header=0)
rooftop_PV = pd.read_csv(path+rooftopPV_file,header=0)

Demand = fix_data(demand).set_index('date_time')
Rooftop_PV = fix_data(rooftop_PV).set_index('date_time')


DATA=Demand.copy()
DATA.rename(columns={'value':'demand'},inplace=True)
DATA['PV'] = Rooftop_PV

DT1 = '2025-10-10 22:00:00'
DT2 = '2025-10-12 06:00:00'


DATA_SA = DATA.loc[DT1:DT2]



import plotly.graph_objects as go
Fig = go.Figure()

for col in ['demand','PV']:
    Fig.add_trace(go.Scatter(x=DATA_SA.index,
                             y=DATA_SA[col],
                             line_shape = 'hv',
                             line = dict(width=1),
                             name=col,
                             stackgroup='one'
                            ))


Fig.show()
```
